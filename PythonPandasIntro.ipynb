```python
# Filtering - output a boolean for each value whether the filter is true or not
# This outputs a "mask" - a Series of the same size as the column with a boolean at each index
female_mask = (df['sex'] == 'female')

female_mask[:10]
```

```python
# Combining masks - *we use & operator for this in pandas*
```

```python
female_kid_mask = (df['sex'] == 'female') & (df['age'] < 25)

female_kid_mask[:10]
```

```python
# Index the original dataframe with the mask
df[female_kid_mask].head()
```

```python
male_old_150p_mask = (df.sex == 'male') & (df.age > 25) & (df.fare > 150)

df[male_old_150p_mask].head()
```

```python
male_older_150p_mask = male_old_150p_mask & (df.age < 75) & (df.age > 50)

df[male_older_150p_mask].head()
```

```python
# Note parenthesis is necessary for OR filtering
female_splitage = (df.sex == 'female') & ((df.age < 25) | (df.age > 75))

df[female_splitage].head()
```

```python
# Built in functions for determining statistics
# min, max, std
df.age.min(), df.age.max(), df.age.std()
```

```python
# Mean ticket price for men and for women
print df[female_mask].fare.mean()

# Not mask for flipping data results - NOTE we use "~"
df[~female_mask].fare.mean()
```

```python
# To select a range of columns, your inner argument needs to be a list of columns
df[['age', 'sibsp', 'parch']].mean()
```

```python
# Applying a custom function to a series
def age_today(age):
    age_today = (2018-1911) + age
    return age_today
```

```python
df['age'].head()
```

```python
# Default axis argument of apply is axis = 0 - i.e. by default, we are feeding the function to every single row
# axis = 1 would try to loop over every column
df['age'].apply(age_today).head()
```

```python
# adding a new column to our original dataframe
# we could also just create a variable for this that would be saved as a series
df['age_today'] = df['age'].apply(age_today)
```

```python
df.head()
```

```python
# Applying a function to a row (as opposed to a series)
# Custom function to sum family size
def get_groupsize(row):
    family = row.sibsp + row.parch
    return 1 + family
```

```python
# Access particular rows using ".loc" then indexing
row_sample = df.loc[1]
row_sample
```

```python
get_groupsize(row_sample)
```

```python
# To apply function to rows, need to change the axis of df.apply to 1 to have it iterate through rows instead
df['groupsize'] = df.apply(get_groupsize, axis=1)
df.head()
```

```python
def travel_trajectory(row):
    trajectory = str(row['embarked']) + str(row['home.dest'])
    return trajectory
```

```python
df['trajectory'] = df.apply(travel_trajectory, axis=1)
df.head()
```

```python
# Sorting values by columns
# "sort.values" by default sorts in ASCENDING order
df.sort_values('fare').head()
```

```python
# can pass in an ascending argument to sort_values to have it sort in descending order
df.sort_values('fare', ascending=False).head()
```

```python
# Can sort by multiple parameters by passing them in as a list
# With multiple parameters, will sort from left to right. 
# In this case, will FIRST sort by pclass, then WITHIN pclass, will sort by fare.
# Can also pass additional parameters that are different. Corresponds to position within list
df.sort_values(['pclass','fare'], ascending=[True, False]).head()
```
